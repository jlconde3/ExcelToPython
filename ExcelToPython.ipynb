```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data_frame = pd.read_excel("Horas Proyecto (Web) - Responsable_3qvswiui.xlsx", 1)
```

```python
data_frame_mod = data_frame.copy()
```

```python
data_frame_mod = data_frame_mod.drop(columns=['T','Compañía', 'Imputador', 'Resno', 'Resno (T)','Proyecto (T)'])
```

```python
data_frame_mod.sort_values('Fcha.asto.')
data_frame_mod.dropna(inplace= True)
data_frame_mod.reset_index()
```

```python
df_group_count = data_frame_mod.groupby(['Fcha.asto.']).count()
df_group_sum = data_frame_mod.groupby(['Fcha.asto.']).sum()
```
